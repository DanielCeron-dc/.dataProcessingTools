## Polynomial regression 

## import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

# Importing the dataset
dataset = pd.read_csv(os.path.join(os.path.abspath(''), 'Position_Salaries.csv')); 

levels = dataset.iloc[:, 1].values; 
salaries = dataset.iloc[:, -1].values;

plt.scatter(levels, salaries, color = 'red');
plt.title('Salary vs Level');


```

```python
#from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

#X_train, X_test, y_train, y_test = train_test_split(levels, salaries, test_size = 0.2, random_state = 0)

linear_regressor = LinearRegression();
linear_regressor.fit(levels.reshape(-1, 1), salaries);

poly_regressor = PolynomialFeatures(degree = 4);
leveles_poly = poly_regressor.fit_transform(levels.reshape(-1, 1));
print(leveles_poly);


linear_regressor2 = LinearRegression();
linear_regressor2.fit(leveles_poly, salaries);

X_grid = np.arange(min(levels), max(levels), 0.1);
X_grid = X_grid.reshape((len(X_grid), 1));


plt.plot(X_grid, linear_regressor.predict(X_grid.reshape(-1, 1)), color = 'blue');
plt.plot(X_grid, linear_regressor2.predict(poly_regressor.fit_transform(X_grid.reshape(-1, 1))), color = 'green');
plt.scatter(levels, salaries, color = 'red');




```

```python
#predition 6.5

print(linear_regressor.predict([[6.5]]))

print(linear_regressor2.predict(poly_regressor.fit_transform([[6.5]])))
```
