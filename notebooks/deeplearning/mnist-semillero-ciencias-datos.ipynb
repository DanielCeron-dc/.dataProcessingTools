# **Fuente de Datos**

MNIST Se trata de un conjunto de datos de 60.000 imgenes en escala de grises de 28x28 de 10 dgitos, junto con un conjunto de prueba de 10.000 imgenes

```python
mnist = tf.keras.datasets.mnist
```

Leer los datos desde la fuente de datos, la fuente nos entrega en este cado un dataset de entrnamiento y prueba

```python
(x_train, y_train),(x_test, y_test) = mnist.load_data()
```

# **Procesamiento de Datos**

Una imagen digital se representa usualmente con una matriz de pixeles, donde cada pixel es un punto de la imagen y tiene un determinado color, en este caso tenemos una imagen en escala de grises, tambin denominada imagen de intensidad, es una matriz con valores enteros entre 0 (Negro) y 255(Blanco)

Los valores de la escala de grises tambin se pueden medir como porcentajes de cobertura de la tinta negra (0 % es igual a blanco, 100 % a negro)

La normalizacin consiste en llevar a un tamao estndar las dimensiones de una imagen, sin provocar en ella alguna distorsin de importancia para esto se divide la matriz entre 255

```python

x_train, x_test = x_train / 255.0, x_test / 255.0
```

Seleccionamos aleatoriamente una imagen y la visualizamos para entender los datos de entrada, tambien se imprime la etiqueta asociada

```python
pos=rnd.randint(0, len(x_train))
img=x_train[pos]
plt.imshow(img, cmap = "gray");
print('Eqtiqueta:',y_train[pos])
```

# **Construcin de modelo**

En una red secuencial los datos se alimentan a la capa de entrada y son procesadas capa por capa (hacia adelante) hasta la capa de salida, la cual tienen como salida un vector, en donde la posicin con el mayor valor indica la clase reconocida.

Las capas Dense son las capas de la red neuronal artificial donde se especifican el nmero de neuronas y la funcin de activacin de cada neurona que la conforma.

Las capas Droput son capas que habilitan o deshabilitan las conexiones entre capas con una probabilidad, esto se utiliza para evitar el sobreentrenamiento de la red

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),

  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10, activation='softmax')
])
```

Se configura los parametros de entrenamiento, los ms importantes:
El algoritmo de entrenamiento
Funcin de perdida o error
Metricas que se utilizarn para verificar el entrenamiento del modelo

```python
model.compile(optimizer='adam',
  loss='sparse_categorical_crossentropy',
  metrics=['accuracy'])
```

# **Entrenar el Modelo**

```python

history=model.fit(x_train, y_train, epochs=5, validation_split=0.1)
```

# **Evaluar el modelo**

```python
model.evaluate(x_test, y_test)
```
